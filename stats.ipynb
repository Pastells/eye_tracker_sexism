```python
import numpy as np
```

# One sample t-test

```python
n = 10_000
a = np.random.normal(5, size=n)

t = (a.mean() - 5) / (a.std(ddof=1)/np.sqrt(n))
t
```

# Two-sample t-test

```python
b = np.random.normal(5, size=n)

sp = np.sqrt((a.var(ddof=1) + b.var(ddof=1))/2)
t = (a.mean()-b.mean())/(sp*np.sqrt(2/n))
t
```
